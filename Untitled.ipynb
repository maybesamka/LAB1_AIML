```python
import pandas as pd
df = pd.read_csv("titanic.csv")
#2 Вывести с помощью python данные из датасета на экран
df.head(10)
```

```python
df.info()
```

```python
#3 Получить количество пропущенных значений для каждого столбца в датасетах
nan_matrix = df.isnull()
print(nan_matrix.sum())
```

```python
#4 Заполнить пропущенные значения в датасете модой/медианой/средним значением и показать, что они действительно были заполнены

df["Age"] = df["Age"].fillna(df["Age"].mode()[0])
df["Cabin"] = df["Cabin"].fillna("Unknown")
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
nan_matrix = df.isnull()
print(nan_matrix.sum())
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler\

a = ['Age', 'Fare', 'SibSp', 'Parch']
df[a] = StandardScaler().fit_transform(df[a])

print(df[a].describe())
```

```python
#Провести преобразование категориальных данных так, чтобы в будущем это не привело к переобучению модели машинного обучения и датасет имел удобоваримый вид

categorical_cols = df.select_dtypes(include=['object']).columns
print("categorical:)
print(categorical_cols)

ohe_cols = ['Sex', 'Embarked', 'Cabin']
drop_cols = ['Name', 'Ticket']

df['Cabin'] = df['Cabin'].str[0] 
df['Cabin'] = df['Cabin'].fillna('Unknown')

df = pd.get_dummies(df, columns=ohe_cols, drop_first=True, prefix=ohe_cols)

df.drop(drop_cols, axis=1, inplace=True, errors='ignore')

print(df.head())
```
